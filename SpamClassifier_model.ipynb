```python
# Importing essential libraries
import pandas as pd
import pickle
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/spam.csv', encoding='latin-1')[['v1', 'v2']]
df.columns = ['label', 'message']
df.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
import string
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
[punc for punc in string.punctuation]
```

```python
def text_process(msg):
    nopunc =[char for char in msg if char not in string.punctuation]
    nopunc=''.join(nopunc)
    return ' '.join([word for word in nopunc.split() if word.lower() not in stopwords.words('english')])
```

```python
df['tokenized_message'] = df['message'].apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(max_df = 0.9,min_df = 10)
X = vectorizer.fit_transform(df['tokenized_message'])
X.toarray()
```

```python
# Extracting dependent variable from the dataset
y = pd.get_dummies(df['label'])
y = y.iloc[:, 1].values
```

```python
# Creating a pickle file for the CountVectorizer
pickle.dump(vectorizer, open('X-transform.pkl', 'wb'))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.2,random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_model = MultinomialNB().fit(X_train,Y_train)
```

```python
# Creating a pickle file for the Multinomial Naive Bayes model
filename = 'spam-sms-model.pkl'
pickle.dump(spam_model, open(filename, 'wb'))
```
